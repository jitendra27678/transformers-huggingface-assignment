The selected model is **bert-base-uncased**, one of the most widely used Transformer-based language models developed by Google. It is based on the encoder component of the Transformer architecture and is designed to learn deep bidirectional representations of text.

- BookCorpus consisting of 11,038 unpublished books.

### Limitations and Biases
Although the training data is relatively neutral, the model can learn and reproduce social biases present in the data. For example, it may associate certain professions more strongly with men or women, leading to biased predictions.

# Introduction
Natural Language Processing (NLP) is a branch of Artificial Intelligence that enables computers to understand, interpret, and generate human language. NLP is widely used in applications such as chatbots, sentiment analysis, machine translation, text summarization, and question-answering systems.

Before Transformers, Recurrent Neural Networks (RNNs) and Long Short-Term Memory (LSTM) networks were commonly used for NLP tasks. However, these models process text sequentially, making training slow and limiting their ability to capture long-range dependencies effectively.

To overcome these limitations, the Transformer architecture was introduced in the landmark paper "Attention Is All You Need" by Vaswani et al. in 2017. Transformers rely on self-attention mechanisms, enabling parallel processing of text and significantly improving performance across a wide range of NLP tasks.

The introduction of Transformers revolutionized NLP and led to the development of powerful models such as BERT, GPT, T5, and BART.

# Transformer Architecture

## Input Embeddings

Transformers cannot directly understand text. Therefore, each word is converted into a numerical vector called an embedding. These embeddings capture semantic meaning and relationships between words.

For example:

"I love NLP"

↓

I      → Vector 1
love   → Vector 2
NLP    → Vector 3

These vectors become the input to the Transformer model.



## Positional Encoding

Unlike RNNs, Transformers process all words simultaneously. Therefore, they need information about word order. Positional encoding is added to input embeddings to provide information about the position of each word in the sentence.

Example:

I      → Position 0
love   → Position 1
NLP    → Position 2

This helps the model understand sentence structure.




## Self-Attention

Self-attention allows each word to focus on other important words in the sentence.

For each word, three vectors are generated:

- Query (Q)
- Key (K)
- Value (V)

The attention mechanism calculates how much focus should be given to every other word in the sentence.
This formula computes attention scores and determines the importance of each word relative to others.

## Multi-Head Attention

Instead of performing a single attention operation, Transformers use multiple attention heads.

Each head learns different linguistic relationships such as:

- Grammar
- Context
- Meaning
- Dependencies

The outputs of all heads are combined to produce richer representations.

## Feed Forward Network (FFN)

The output from the attention layer is passed through a fully connected neural network called the Feed Forward Network.

Structure:

Linear Layer
↓
ReLU Activation
↓
Linear Layer

This helps the model learn complex patterns and representations.

## Residual Connections

Residual connections help prevent information loss during training.

Output = Layer(x) + x

These connections improve gradient flow and make deep networks easier to train.

## Layer Normalization

Layer normalization stabilizes the learning process by normalizing activations.

Benefits:

- Faster convergence
- Stable training
- Better performance


## Encoder Architecture

The encoder consists of multiple identical layers.

Each encoder layer contains:

1. Multi-Head Attention
2. Residual Connection
3. Layer Normalization
4. Feed Forward Network
5. Residual Connection
6. Layer Normalization

BERT is based entirely on the encoder architecture.

## Decoder Architecture

The decoder also consists of multiple layers and includes:

1. Masked Multi-Head Attention
2. Encoder-Decoder Attention
3. Feed Forward Network

The decoder generates output tokens one by one.

Models such as GPT primarily use decoder-based architectures.



### Transformer Flow

Input Sentence

↓

Input Embeddings

↓

Positional Encoding

↓

Encoder

↓

Decoder

↓

Output Sentence

# Practical Implementation

## Task 1: Text Summarization using AutoClasses

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

model_name = "facebook/bart-large-cnn"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

text = """
Transformers have revolutionized natural language processing by enabling parallel processing and improved contextual understanding. Unlike traditional recurrent neural networks, Transformers rely on self-attention mechanisms that allow them to process all words in a sentence simultaneously. This innovation has led to significant improvements in machine translation, text summarization, question answering, and many other NLP tasks.
"""

inputs = tokenizer(
    text,
    return_tensors="pt",
    max_length=1024,
    truncation=True
)

summary_ids = model.generate(
    inputs["input_ids"],
    max_length=50,
    min_length=20,
    num_beams=4
)

summary = tokenizer.decode(
    summary_ids[0],
    skip_special_tokens=True
)

print("Summary:")
print(summary)
```

## Task 1: Text Summarization Results

### Input Text

Transformers have revolutionized natural language processing by enabling parallel processing and improved contextual understanding. Unlike traditional recurrent neural networks, Transformers rely on self-attention mechanisms that allow them to process all words in a sentence simultaneously. This innovation has led to significant improvements in machine translation, text summarization, question answering, and many other NLP tasks.

### Generated Summary

Transformers have revolutionized natural language processing. Unlike traditional recurrent neural networks, Transformers rely on self-attention mechanisms. This has led to significant improvements in machine translation, text summarization, and question answering.

### Observations

The BART model successfully condensed the original paragraph into a shorter summary while preserving the key information. The generated summary is coherent, grammatically correct, and captures the main idea of the input text.

## Task 2: Language Translation using AutoClasses

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

model_name = "Helsinki-NLP/opus-mt-en-fr"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

text = "Artificial Intelligence is transforming the world."

inputs = tokenizer(text, return_tensors="pt")

translated = model.generate(**inputs)

output = tokenizer.decode(
    translated[0],
    skip_special_tokens=True
)

print("Translation:")
print(output)
```

## Task 2: Language Translation Results

### Input Text

Artificial Intelligence is transforming the world.

### Generated Output

L'intelligence artificielle transforme le monde.

### Observations

The MarianMT model successfully translated the English sentence into French while preserving the original meaning. The translated sentence is grammatically correct and demonstrates the effectiveness of Transformer-based machine translation models.
